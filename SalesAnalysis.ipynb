```python
all_data = pd.read_csv(r"c:\\Users\\user\\AppData\\Local\\Programs\\Microsoft VS Code/all_data.csv")
all_data.head()
```

### find 'or' and delete it

```python
all_data = all_data[all_data['Order Date'].str[0:2]!='Or']
all_data.head()
```

```python
'''
Now that we have remove the columns with the 'Or' 
we can continue with convering month column to an int
'''
```

```python
all_data = all_data.dropna(how='all')
```

### TASK 1: adding a month column

```python
all_data['Month'] = all_data['Order Date'].str[0:2]
all_data['Month'] = all_data['Month'].astype('int32')
all_data.head()
```

### TASK 2: Adding a sales column

```python
#converting these columns into int and float
all_data['Quantity Ordered'] = all_data['Quantity Ordered'].astype('int32')
all_data['Price Each'] = all_data['Price Each'].astype('float')
all_data.head()
```

```python
all_data['Sales'] = all_data['Quantity Ordered'] * all_data['Price Each']
all_data.head()
```

### Question: Which month had the best sales?

```python
results = all_data.groupby('Month').sum()
results.head()
```

```python
#visualisation import matplotlib
import matplotlib.pyplot as plt

months = range(1,13)


plt.bar(months, results['Sales'])
plt.show()
```

#### TASK 5: What products are most often sold together?

```python
new_all = all_data[all_data['Order ID'].duplicated(keep=False)]
new_all.head(20)
```

```python
#bundling two or more rows into one. Using the transform() function
#Joining a few products with the same order ID into the same line.
new_all['Product_Bundle'] = new_all.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))

#Dropping the duplicate values
new_all = new_all[['Order ID', 'Product_Bundle']].drop_duplicates()

new_all.head()
```

```python
#importing libraries
from itertools import combinations
from collections import Counter
```

```python
#creating an empty Count()

count = Counter()

for row in new_all['Product_Bundle']:
    row_list = row.split(',')
    #count.update(Counter(combinations(row_list, 2)))
    count.update(Counter(combinations(row_list, 3)))

count.most_common(10)
```

```python
product_group = all_data.groupby('Product')
quantity_ordered = all_data.groupby('Product').sum()['Quantity Ordered']

products = [product for product, df in product_group]

plt.bar(products, quantity_ordered)
plt.ylabel('Quantity Ordered')
plt.xlabel('Product')
plt.xticks(rotation=90)
plt.show()
```

```python
#overlaying a second y-axis on existing chart

prices = all_data.groupby('Product').mean()['Price Each']
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(products, quantity_ordered, color='g')
ax2.plot(products, prices, 'r-')

ax1.set_xlabel('Product Name')
ax1.set_ylabel('Quantity Ordered', color='g')
ax2.set_ylabel('Price ($)', color='b')
ax1.set_xticklabels(products, rotation='vertical', size=8)

plt.show()
```
